**Задание 5. В двух столбцах есть пропущенные значения. Один из них — `days_employed`. Пропуски в этом столбце вы обработаете на следующем этапе. Другой столбец с пропущенными значениями — `total_income` — хранит данные о доходах. На сумму дохода сильнее всего влияет тип занятости, поэтому заполнить пропуски в этом столбце нужно медианным значением по каждому типу из столбца `income_type`. Например, у человека с типом занятости `сотрудник` пропуск в столбце `total_income` должен быть заполнен медианным доходом среди всех записей с тем же типом.**

```python
for t in data['income_type'].unique():
    data.loc[(data['income_type'] == t) & (data['total_income'].isna()), 'total_income'] = \
    data.loc[(data['income_type'] == t), 'total_income'].median()
```

### Категоризация данных

**Задание 16. На основании диапазонов, указанных ниже, создайте в датафрейме `data` столбец `total_income_category` с категориями:**

- 0–30000 — `'E'`;
- 30001–50000 — `'D'`;
- 50001–200000 — `'C'`;
- 200001–1000000 — `'B'`;
- 1000001 и выше — `'A'`.


**Например, кредитополучателю с доходом 25000 нужно назначить категорию `'E'`, а клиенту, получающему 235000, — `'B'`. Используйте собственную функцию с именем `categorize_income()` и метод `apply()`.**

```python
def categorize_income(income):
    try:
        if 0 <= income <= 30000:
            return 'E'
        elif 30001 <= income <= 50000:
            return 'D'
        elif 50001 <= income <= 80000:
            return 'C3'
        elif 80001 <= income <= 130000:
            return 'C2'
        elif 130001 <= income <= 200000:
            return 'C1'
        elif 200001 <= income <= 350000:
            return 'B3'
        elif 350001 <= income <= 600000:
            return 'B2'
        elif 600001 <= income <= 1000000:
            return 'B1'
        elif income >= 1000001:
            return 'A'
    except:
        pass
```

```python
data['total_income_category'] = data['total_income'].apply(categorize_income)
```

**Задание 18. Создайте функцию, которая на основании данных из столбца `purpose` сформирует новый столбец `purpose_category`, в который войдут следующие категории:**

- `'операции с автомобилем'`,
- `'операции с недвижимостью'`,
- `'проведение свадьбы'`,
- `'получение образования'`.

**Например, если в столбце `purpose` находится подстрока `'на покупку автомобиля'`, то в столбце `purpose_category` должна появиться строка `'операции с автомобилем'`.**

**Используйте собственную функцию с именем `categorize_purpose()` и метод `apply()`. Изучите данные в столбце `purpose` и определите, какие подстроки помогут вам правильно определить категорию.**

```python
def categorize_purpose(row):
    try:
        if 'автом' in row:
            return 'операции с автомобилем'
        elif 'жил' in row or 'недвиж' in row:
            return 'операции с недвижимостью'
        elif 'свад' in row:
            return 'проведение свадьбы'
        elif 'образов' in row:
            return 'получение образования'
    except:
        return 'нет категории'
```

```python
data['purpose_category'] = data['purpose'].apply(categorize_purpose)
```

#### 3.1 Есть ли зависимость между количеством детей и возвратом кредита в срок?

Чтобы понять существует ли зависимость между количеством детей и возвратом кредита в срок, для начала сгруппируем данные по количеству детей и посчитаем сумму задолженности по каждой группе. 


```python
# Сгруппируем данные по количеству детей и посчитаем сумму задолженности по каждой группе. 
# Сумму по каждой группе сохраним в переменную children_debt_sum, вдруг понадобятся еще действия с ними.
children_debt_sum = data.groupby('children')['debt'].sum()
display(children_debt_sum)# Ваш код будет здесь. Вы можете создавать новые ячейки.
```

Получается что больше всего должников бездетные, а при увеличении количества детей возвращаемость кредитов увеличивается. 

Но данные показатели нельзя считать объективными так как количество заемщиков в каждой группе разное.


```python
# Узнаем количество заемщиков в каждой группе. 
# Количество заемщиков сохраним в переменную children_debt_count
children_debt_count = data.groupby('children')['debt'].count()
display(children_debt_count)
```

Узнав количество заемщиков. Получается что больше всего заемщиков бездетные, а заемщиков с количеством детей больше 5 всего 9. 

Имея количество заемщиков по каждой группе и количество должников по каждой группе. Можно узнать отношение должников к общему количеству заемщиков по каждой группе и их процентное выражение.

```python
# Узнаем отношение должников к общему количеству заемщиков по каждой группе.
ratio_dept_children = (children_debt_sum / children_debt_count)
display('Процент должников каждой группы равен:', ratio_dept_children * 100)
```

Получается что в группе заемщиков без детей меньше всего должников только 7,5 процентов, а при увеличении количества детей в семье увеличивается число должников по кредиту от 8.18% до 9.75%. Хотя процент должников в семьях с 3 детьми меньше чем в семьях с одним или двумя детьми. А семьи с пятью детьми вообще не имеют задолженности когда в среди семей с четырьмя детьми самый большой процент должников 9.75, скорее всего это связано с недостаточным количеством, для подобного анализа, заемщиков с пятью детьми в семье.

Решение можно было упростить определив сразу среднее значение, в одну строку, создав сводную таблицу

```python
data_pivot = data.pivot_table(index='children', values='debt', aggfunc=['count', 'sum', 'mean'])
display(data_pivot)
```

Можно просмотреть средние значения соседних групп семей, чтобы снивелировать значения.

```python
print(data_pivot.loc[0:1, 'mean'].mean()*100)
print(data_pivot.loc[1:2, 'mean'].mean()*100)
print(data_pivot.loc[2:3, 'mean'].mean()*100)
print(data_pivot.loc[3:4, 'mean'].mean()*100)
print(data_pivot.loc[4:5, 'mean'].mean()*100)
```

 В любом случае учитывать данные по семьям с пятью детьми в данном случае нежелательно, так как в данном, конкретном случае девять семей с пятью детьми, это мало для анализа. Для определения всех возрастов людей сгруппируем их по возрастанию и посчитаем количество каждого возраста.

```python
display(data.groupby('dob_years')['dob_years'].count())
```

При этом выявилась аномалия, у 100 человек возраст 0 чего быть не может. Так как количество аномалий не большое меньше половины процента, нулевые значения можно заменить медианными значениями.

```python
data.loc[data['dob_years'] == 0, 'dob_years'] = data['dob_years'].median()
```

Проверим сколько нулевых значений осталось в столбце.

```python
print(data.loc[data['dob_years'] == 0, 'dob_years'].count())
```

Далее необходимо категоризировать возрастные диапазоны с условием
- 19–30 — `'молодые'`;
- 31–47 — `'средний возраст'`;
- 48–65 — `'пожилые'`;
- старше 65 — `'пенсионеры'`.

```python
def categorize_age (year):
    if year <= 30:
        return 'молодые'
    if year <= 47:
        return 'средний возраст'
    if year <= 65:
        return 'пожилые'
    return 'пенсионеры'
```

Добавим в исходный датафрейм столбец **categorize_years** с категориями соответствующими возрасту заемщика методом **apply()**.

```python
data['categorize_years'] = data['dob_years'].apply(categorize_age)
```

Ну и для определения зависимости возраста плательщика на возврат кредита в срок составим сводную таблицу в которой сгруппируем данные по столбцу **categorize_years**, который добавили в таблицу data. В таблицу поместим значения задолженности из столбца **debt** и определим их *количество*, *сумму* и *среднее значение*.

```python
data_pivot = data.pivot_table(index='categorize_years', values='debt', aggfunc=['count', 'sum', 'mean'])
display(data_pivot)
```

**Вывод** Исходя из полученных данных видно что больше всего заемщиков в возрасте от 30 до 47 лет. Меньше всего к просрочке или невыплате платежа склонны люди в возрасте от 48  и старше. Чаще  всего, иметь долги по кредиту, могут себе люди в возрасте от 19 до 30 лет. Возраст явно влияет на выплату кредита. С увеличением возраста, увеличивается и ответственность по выплате кредита.
